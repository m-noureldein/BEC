```python
# Extract sequences from fasta
# Import modules
from Bio import SeqIO

filename='Downloads/rosalind_long-6.txt'
sequences=[i for i in SeqIO.parse(filename, 'fasta')]

interests=[]

for i in range(0, len(sequences)):
    interests.append(str(sequences[i].seq))
```

```python
interests=[]

for i in range(0, len(sequences)):
    interests.append(str(sequences[i].seq))
    
comb=str(interests[0])

x=int(0.5*len(interests[0]))


for i in range(0,50):
    for j in range(0,len(interests)):
        if interests[j] not in comb:
            if comb[-x:] in interests[j]:
                h=comb[-x:]
                comb+=interests[j].split(h,1)[1]
            elif comb[:x] in interests[j]:
                h=comb[0:x]
                comb=interests[j].split(h,1)[0]+comb


comb
```
