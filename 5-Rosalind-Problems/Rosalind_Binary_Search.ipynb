# Binary Search
<https://rosalind.info/problems/bins/>

```python
with open('Downloads/rosalind_bins-4.txt', 'r') as authfile:
    n = int(authfile.readline().rstrip())
    m = int(authfile.readline().rstrip())
    A=[int(i) for i in (authfile.readline().rstrip()).split()]
    B=[int(i) for i in (authfile.readline().rstrip()).split()]
x=round(n/2)
```

```python
%%time
ind=[]
for i in range(0, len(B)):
    if B[i] in A[0:x]:
        for j in range(0, x):
            if B[i]==A[j]:
                ind.append(j+1)
                break
    elif B[i] in A[x:n]:
        for j in range(x,n):
            if B[i]==A[j]:
                ind.append(j+1)
                break
    else:
        ind.append(-1)
        
with open('Downloads/readme.txt', 'w') as f:
    f.write(' '.join(map(str,ind)))
```
