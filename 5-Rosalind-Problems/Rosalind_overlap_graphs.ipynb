# Overlap Graphs Problem
<https://rosalind.info/problems/grph/>

```python
# Extract sequences from fasta
# Import modules
from Bio import SeqIO
```

```python
filename='Downloads/rosalind_grph-4.txt'
sequences=[i for i in SeqIO.parse(filename, 'fasta')]
```

```python
for i in range(0, len(sequences)):
    for j in range(0, len(sequences)):
        if sequences[i].seq[-3:]==sequences[j].seq[0:3]:
            if sequences[i].name != sequences[j].name:
                print(sequences[i].name+" "+sequences[j].name)
    
```
