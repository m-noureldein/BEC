# Error Correction in Reads
<https://rosalind.info/problems/corr/>

```python
# Extract sequences from fasta
# Import modules
from Bio import SeqIO

filename='Downloads/rosalind_corr-2.txt'
sequences=[i for i in SeqIO.parse(filename, 'fasta')]

interests=[]

for i in range(0, len(sequences)):
    interests.append(str(sequences[i].seq))
    
len(interests)
```

```python
DNA_ReverseComplement={'A':'T','T':'A','G':'C','C':'G'}
def reverse_complement(seq):
    """Swapping adenine with thymine and guanine with cytosine. Reversing newly generated string"""
    return ''.join([DNA_ReverseComplement[nuc] for nuc in seq])[::-1]
```

```python
err=[]
ress=[]
for i in range(0, len(interests)):
    temp=0
    for j in range(0, len(interests)):
        if interests[i]==interests[j]:
            temp+=1
        elif reverse_complement(interests[i])==interests[j]:
            temp+=1
    if temp==1:
        err.append(interests[i])
    else:
        ress.append(interests[i])
        
```

```python
ress2=[]
for i in range(0, len(ress)):
    ress2.append(reverse_complement(ress[i]))
    
ress3=ress+ress2
    
```

```python
final_seq=[]
for i in range(0, len(err)):
    for k in range(0, len(ress3)):
        temp=0
        for j in range(0, len(err[i])):
            if err[i][j]==ress3[k][j]:
                temp+=1
            if temp==(len(err[i])-1):
                temp2=err[i]+"->"+ress3[k]
                if temp2 not in final_seq:
                    final_seq.append(temp2) 

        
print('\n'.join(map(str, final_seq)))          
            
            
```
